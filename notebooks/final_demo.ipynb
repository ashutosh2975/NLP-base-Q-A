```python
import re

def parse_tags(tag_str):
    if pd.isna(tag_str):
        return []
    return re.findall(r"'([^']+)'", tag_str)
```

```python
sample = df.sample(5, random_state=42)

for _, row in sample.iterrows():
    print("\nQUESTION:")
    print(row["Processed_Text"][:150])
    print("ORIGINAL TAGS:", parse_tags(row["Tags_List"]))
```
